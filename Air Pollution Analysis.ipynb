```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pollutants = ['PM2.5', 'PM10', 'NO2', 'CO', 'SO2', 'O3']
def visualize_air_quality_distribution_stats(data_cleaned, pollutants):
    while True:
        print("\nAvailable pollutants to visualize:")
        for idx, pollutant in enumerate(pollutants, start=1):
            print(f"{idx}. {pollutant}")

        try:
            choice_pollutant = int(input(f"\nEnter the number corresponding to the pollutant you want to visualize (1-{len(pollutants)}): "))
            if choice_pollutant < 1 or choice_pollutant > len(pollutants):
                print("Invalid choice. Please select a valid number.")
                continue

            selected_pollutant = pollutants[choice_pollutant - 1]
            print("\nAvailable statistics:")
            print("1. Mean")
            print("2. Median")
            print("3. Standard Deviation")
            print("4. Maximum")
            print("5. Minimum")

            choice_stat = int(input(f"Enter the number corresponding to the statistic you want to use (1-5): "))
            if choice_stat < 1 or choice_stat > 5:
                print("Invalid choice. Please select a valid number.")
                continue
            stat_functions = {
                1: 'mean',
                2: 'median',
                3: 'std',
                4: 'max',
                5: 'min'
            }
            stat_func = stat_functions[choice_stat]
            stat_value = getattr(data_cleaned[selected_pollutant], stat_func)()
            print(f"\n{stat_func.capitalize()} of {selected_pollutant}: {stat_value}")
            plt.figure(figsize=(10, 6))
            sns.histplot(data_cleaned[selected_pollutant], bins=30, kde=True, color='blue')
            plt.title(f'Distribution of {selected_pollutant} (Statistic: {stat_func.capitalize()})', fontsize=14)
            plt.xlabel(f'{selected_pollutant} Levels', fontsize=12)
            plt.ylabel('Frequency', fontsize=12)
            plt.tight_layout()
            plt.show()

        except ValueError:
            print("Invalid input. Please enter a valid number.")

        # Ask the user if they want to continue or exit
        continue_choice = input("\nDo you want to analyze another pollutant? (yes/no): ").strip().lower()
        if continue_choice != 'yes':
            print("Exiting the program. Goodbye!")
            break
data = {
    'PM2.5': [12, 15, 18, 13, 14, 20, 21],
    'PM10': [30, 35, 40, 45, 50, 55, 60],
    'NO2': [30, 25, 20, 22, 28, 35, 40],
    'CO': [0.5, 0.8, 0.7, 0.9, 1.0, 1.1, 1.2],
    'SO2': [0.03, 0.02, 0.04, 0.03, 0.05, 0.06, 0.07],
    'O3': [100, 110, 120, 130, 125, 115, 105]
}

data_cleaned = pd.DataFrame(data)
visualize_air_quality_distribution_stats(data_cleaned, pollutants)
```

```python
import pandas as pd
data = {
    'City': ['Ahmedabad', 'Aizwal', 'Amaravathi', 'Amritsar', 'Bengaluru', 'Bhopal', 'Brajrajnagar', 'Chandigarh', 'Chennai', 'Coimbatore',],
    'AQI': [50, 120, 150, 180, 90, 200, 60, 75, 300, 400]
}
data_cleaned = pd.DataFrame(data)
bins = [0, 50, 100, 150, 200, 300, 500]
labels = ['good', 'moderate', 'satisfactory', 'poor', 'very poor', 'severe']
data_cleaned['AQI_Bucket'] = pd.cut(data_cleaned['AQI'], bins=bins, labels=labels, right=False)
def show_cities_in_aqi_bucket(data_cleaned, labels):
    while True:
        print("\nAvailable AQI Buckets:")
        for idx, label in enumerate(labels, start=1):
            print(f"{idx}. {label}")
        
        try:
            choice_bucket = int(input(f"\nEnter the number corresponding to the AQI bucket (1-{len(labels)}): "))
            if choice_bucket < 1 or choice_bucket > len(labels):
                print("Invalid choice. Please select a valid number.")
                continue
            
            selected_bucket = labels[choice_bucket - 1]
            cities_in_bucket = data_cleaned[data_cleaned['AQI_Bucket'] == selected_bucket]['City']
            if not cities_in_bucket.empty:
                print(f"\nCities with AQI bucket '{selected_bucket}':")
                for city in cities_in_bucket:
                    print(city)
            else:
                print(f"No cities found in the '{selected_bucket}' AQI bucket.")
        
        except ValueError:
            print("Invalid input. Please enter a valid number.")
        continue_choice = input("\nDo you want to check another AQI bucket? (yes/no): ").strip().lower()
        if continue_choice != 'yes':
            print("Exiting the program. Goodbye!")
            break
show_cities_in_aqi_bucket(data_cleaned, labels)
```
